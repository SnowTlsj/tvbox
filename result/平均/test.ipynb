# 香氛识别（Jupyter版）

当前版本只保留两部分：
1. 信号处理与40s对齐，并导出新的xlsx到 `result/xlsx`（格式与原始表一致）
2. 图1输出，拆分为4张图片（四季各1张，包含该季全部样本的全部30通道）

```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.size'] = 11

BASE_DIR = Path.cwd()
DATA_DIR = BASE_DIR / 'data'
RESULT_DIR = BASE_DIR / 'result'
XLSX_DIR = RESULT_DIR / 'xlsx'
FIG_DIR = RESULT_DIR / 'fig1_split'

XLSX_DIR.mkdir(parents=True, exist_ok=True)
FIG_DIR.mkdir(parents=True, exist_ok=True)

LABEL_DICT = {'chun': 0, 'xia': 1, 'qiu': 2, 'dong': 3}
CLASS_NAMES = ['春季香氛', '夏季香氛', '秋季香氛', '冬季香氛']
CLASS_KEY = ['chun', 'xia', 'qiu', 'dong']

print(f'数据目录: {DATA_DIR}')
print(f'对齐xlsx输出目录: {XLSX_DIR}')
print(f'图像输出目录: {FIG_DIR}')
```

```python
def detect_injection_point(signal, search_start=35, search_end=55):
    """在给定区间内检测进样点（变化率最大的位置）。"""
    sig = savgol_filter(signal, 11, 2, axis=0)
    diff = np.abs(np.diff(sig, axis=0).mean(axis=1))
    search_zone = diff[search_start:search_end]
    if len(search_zone) == 0:
        return 40
    return int(np.argmax(search_zone) + search_start)


def align_to_40s(signal, real_inj_idx, total_len=200, align_target=40):
    """将真实进样点对齐到第40秒，并返回固定长度信号。"""
    start_idx = real_inj_idx - align_target
    end_idx = start_idx + total_len
    raw_len = len(signal)

    if start_idx < 0:
        pad_left = -start_idx
        part = signal[0:end_idx]
        padding = np.tile(signal[0], (pad_left, 1))
        aligned = np.vstack([padding, part])
    elif end_idx > raw_len:
        pad_right = end_idx - raw_len
        part = signal[start_idx:raw_len]
        padding = np.tile(signal[-1], (pad_right, 1))
        aligned = np.vstack([part, padding])
    else:
        aligned = signal[start_idx:end_idx]

    baseline = aligned[0:10].mean(axis=0)
    return aligned - baseline
```

```python
def load_align_and_export(data_dir=DATA_DIR, out_dir=XLSX_DIR):
    if not data_dir.exists():
        raise FileNotFoundError(f'找不到数据目录: {data_dir}')

    file_list = sorted([p for p in data_dir.iterdir() if p.suffix.lower() in ['.xlsx', '.xls']])
    print(f'读取文件数量: {len(file_list)}')

    X_aligned, y, files = [], [], []
    summary_rows = []

    for fpath in file_list:
        try:
            raw = pd.read_excel(fpath, header=None).values.astype(float)
        except Exception as e:
            print(f'[跳过] {fpath.name} 读取失败: {e}')
            continue

        label = None
        for k, v in LABEL_DICT.items():
            if k in fpath.name.lower():
                label = v
                break
        if label is None:
            print(f'[跳过] {fpath.name} 未匹配到类别')
            continue

        inj_idx = detect_injection_point(raw, search_start=35, search_end=55)
        aligned = align_to_40s(raw, inj_idx, total_len=200, align_target=40)

        X_aligned.append(aligned)
        y.append(label)
        files.append(fpath.name)

        # 关键：按原始格式导出（纯数值、无表头、无索引、单sheet），避免Origin读取异常。
        out_name = f'{fpath.stem}_aligned40.xlsx'
        out_path = out_dir / out_name
        pd.DataFrame(aligned).to_excel(out_path, index=False, header=False)

        summary_rows.append({
            'source_file': fpath.name,
            'class_id': label,
            'class_name': CLASS_NAMES[label],
            'detected_injection_index': inj_idx,
            'output_xlsx': out_name,
            'rows': aligned.shape[0],
            'cols': aligned.shape[1]
        })

    if len(X_aligned) == 0:
        raise ValueError('无有效数据，未生成任何对齐结果。')

    summary_df = pd.DataFrame(summary_rows)
    summary_df.to_excel(out_dir / 'alignment_summary.xlsx', index=False)

    X_aligned = np.array(X_aligned)
    y = np.array(y)

    print(f'完成对齐样本: {len(X_aligned)}')
    print(f'每条信号形状: {X_aligned.shape[1:]} (time, channels)')
    print(f"汇总文件: {out_dir / 'alignment_summary.xlsx'}")

    for i, name in enumerate(CLASS_NAMES):
        print(f'{name}: {(y == i).sum()} 个样本')

    return X_aligned, y, files


X_data, y, files = load_align_and_export()
```

```python
def plot_fig1_split_all_channels(X_data, y, out_dir=FIG_DIR):
    colors = plt.cm.tab10.colors

    for i, class_name in enumerate(CLASS_NAMES):
        subset = X_data[y == i]  # (N, T, C)
        if len(subset) == 0:
            print(f'[跳过] {class_name} 无样本')
            continue

        n_samples, n_time, n_channels = subset.shape
        fig, ax = plt.subplots(figsize=(12, 8))

        # 每个样本的每个通道都绘制，满足“24样本*30通道全部显示”。
        for s in range(n_samples):
            for ch in range(n_channels):
                ax.plot(subset[s, :, ch], color=colors[i], alpha=0.06, linewidth=0.6)

        # 同时加一条季节均值曲线，便于观察总体趋势。
        seasonal_mean = subset.mean(axis=(0, 2))
        ax.plot(seasonal_mean, color='black', linewidth=2.0, label='季节均值(样本×通道)')

        ax.axvline(x=40, color='gray', linestyle=':', alpha=0.9, linewidth=1.2)
        ax.text(41, ax.get_ylim()[1] * 0.95, '进样点 40s', color='gray', fontsize=10)

        ax.set_title(
            f'{class_name} 响应曲线 | 样本: {n_samples} | 通道: {n_channels} | 总线条: {n_samples * n_channels}',
            fontsize=13,
            fontweight='bold'
        )
        ax.set_xlabel('时间 (s)')
        ax.set_ylabel('响应值 (V)')
        ax.grid(alpha=0.25, linestyle=':')
        ax.legend(loc='upper right')

        out_name = f'fig1_{CLASS_KEY[i]}_all_channels.png'
        out_path = out_dir / out_name
        fig.savefig(out_path, dpi=300, bbox_inches='tight')
        plt.close(fig)
        print(f'已保存: {out_path}')


plot_fig1_split_all_channels(X_data, y)
```

```python
print('输出完成。')
print(f'对齐xlsx目录: {XLSX_DIR}')
print(f'图像目录: {FIG_DIR}')
```
